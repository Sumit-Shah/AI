```python
#Checking if there is any kind of dublicate values .
duplicates=data_sydney[data_sydney.duplicated()]
print("Number of duplicate rows:",len(duplicates))
```

```python
#Droping the dublicate value if present in out data set.
data_sydney.drop_duplicates(inplace=True)
```

```python
print("\nSummary Statistics:")
print(data_sydney.describe())
'''displaying  the mean, count, standard deviation, minimum, (Q1), median ( Q2),(Q3), and maximum values for each numerical column in the DataFrame of Weatherdata.'''
```

```python
#Displying  data set   total dataframe .
data_sydney.info(memory_usage='deep')
```

```python
#Checking missing_columns where missing values can be present.
missing_columns=data_sydney.columns[data_sydney.isnull().any()]
missing_columns
#List of missing value that we got .
data_sydney['preciptype'] = data_sydney['preciptype'].fillna('No Data')
data_sydney['snow'] = data_sydney['snow'].fillna('1')
data_sydney['snowdepth'] = data_sydney['snowdepth'].fillna('2')
data_sydney['windgust'] = data_sydney['windgust'].fillna('3')
data_sydney['severerisk'] = data_sydney['severerisk'].fillna('4')
```

```python
#Filling the missing value for our calculation
unique_values = data_sydney['preciptype'].unique()
unique_values = data_sydney['snow'].unique()
unique_values = data_sydney['snowdepth'].unique()
unique_values = data_sydney['windgust'].unique()
unique_values = data_sydney['severerisk'].unique()
print(unique_values)
```

```python
plt.figure(figsize=(20, 10))
plt.plot(data_sydney['temp'], label='Temperature')
plt.plot(data_sydney['humidity'], label='Humidity')
plt.title('Line Chart of Temperature and Humidity')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
#Univariate Analysis-Skewness and Modality:
plt.figure(figsize=(20, 10))
plt.subplot(1, 2, 1)
sns.histplot(data_sydney['temp'], kde=True)
plt.title('Distribution of Temperature (C)')

plt.subplot(1, 2, 2)
sns.histplot(data_sydney['humidity'], kde=True)
plt.title('Distribution of Humidity')

plt.show()
```

```python
import matplotlib.pyplot as plt

#Counting weather  codition for each occurance
condition_counts = data_sydney['conditions'].value_counts()

# Defing the different kind of colour as per our need for piechart
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0', '#ffb3e6']

# Pipe chart from our data .
plt.figure(figsize=(8, 8))
plt.pie(condition_counts, labels=condition_counts.index, autopct='%1.1f%%', startangle=90, colors=colors)

# Seting title
plt.title('Distribution of Weather Conditions')

#Displaying the pi-chart.
plt.show()
```

```python
#import pandas as pd
# For correct analysis seletection only two columns
selected_columns = ['humidity', 'temp']

# Selective columns calculation of corelation matrix.
correlation_matrix = data_sydney[selected_columns].corr()

# Display the correlation matrix
print("\nCorrelation Matrix for Humidity and Temperature Columns:")
print(correlation_matrix)
```

```python
selected_columns = ['humidity', 'temp']
correlation_matrix = data_sydney[selected_columns].corr()

# Using seaborn creating  heatmap.
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap for Humidity and Temperature (C)')
plt.show()
```

Both Humidity and Temperature have a positive linear relation which mean if temperature increase then humidity also increase and vice versa.
